```python
# Synchronization measures
# Summary: Synchrony between clusenters over time for all n_bridges groups n_lines = len(n_bridges list)
# Synchrony between clusters over time for each n_bridges value (all runs withing n_bridges group)
# Synchrony between each cluster pair over time -
# Synchrony of global, cluster and individual io neurons to Ohrstein-Uhlenbeck




# p_bridges /  n_bridges
dat_pre = np.load(p_bridges_filepath, allow_pickle=True)
bridge_dat = {
    cond: dat_pre[cond].item()   # unwrap 0-D object array -> dict
    for cond in ("static", "plasticity")
}

n_bridges_list = np.unique(bridge_dat['plasticity']['n_bridges'])
n_bridge_idx =  { f'{n_bridges}': np.where(bridge_dat['static']['n_bridges'] == n_bridges)[0] for n_bridges in n_bridges_list} # Keys = all possibles values for variable, items = the indices of the runs with those variables


# Calculate phase for all io_signals + their Ohrnstein Uhlenbeck input:
dt = bridge_dat["static"]['dt'][0]
Fs = 1 / (dt / 1000)

io_phases_all= {}
io_amplitudes_all = {}
OU_phases_all = {}
OU_amplitudes_all= {}
d_io_OU_phase= {}

for cond in ("static", "plasticity"):

    # Calculate phases
    io_phases_all[cond], io_amplitudes_all[cond]= filter_and_Hilbert(bridge_dat[cond]['io.V_soma'], Fs)
    OU_phases_all[cond], OU_amplitudes_all[cond]= filter_and_Hilbert( bridge_dat[cond]['io.V_soma'], Fs)


    # Calculate phase difference between IO signal and its I_OU
    d_io_OU_phase[cond] = io_phases_all[cond] - OU_phases_all[cond]


results = nested_dict()

for n_bridges, run_ids  in n_bridge_idx.items():
     for cond in ("static", "plasticity"):
        for weight_use in ('unweighted', 'weighted'):

            n_datapoints = len(io_phases_all[cond][0])
            n_runs = len(run_ids)
            max_n_clusters = np.max(bridge_dat[cond]['IO_n_clusters'])

            bucket = results[cond][weight_use].setdefault(f'{n_bridges}', {
                    "R_K_sync_global": np.zeros((n_datapoints, n_runs)), # Weighted and unweighted Kuramoto synchrony measures
                    "R_K_sync_within": np.full((n_datapoints, max_n_clusters, n_runs), np.nan),
                    "R_K_sync_between": np.zeros((n_datapoints, n_runs)),

                    "global_mean_phase": np.zeros((n_datapoints, n_runs)), # Mean phase overall and within each cluster
                    "within_mean_phase": np.full((n_datapoints, max_n_clusters, n_runs), np.nan),
                    "between_mean_phase": np.zeros((n_datapoints, n_runs)),

                    "OU_sync_global": np.zeros((n_datapoints, n_runs)), # Synchrony to OU on multiple scales
                    "OU_sync_within":  np.full((n_datapoints, max_n_clusters, n_runs), np.nan),
                    "OU_sync_between": np.zeros((n_datapoints, n_runs)),

                    "OU_global_mean_phase": np.zeros((n_datapoints, n_runs)),
                    "OU_within_mean_phase": np.full((n_datapoints, max_n_clusters, n_runs), np.nan),
                    "OU_between_mean_phase": np.zeros((n_datapoints, n_runs))})

            for run_id in run_ids:

                phase_signals = io_phases_all[cond][run_id]
                amplitude_signals = io_amplitudes_all[cond][run_id]
                phase_differences = d_io_OU_phase[cond][run_id]

                # ------------(Un)weighted Kuramoto synchrony ------------
                weights_arg = amplitude_signals if weight_use == "weighted" else None

                R_K_sync_global, global_mean_phase= Kuramoto_order(phase_signals,  weights = weights_arg)

                R_K_sync_within, within_mean_phase, mean_amplitude_clusters = within_cluster_Kuramoto(phase_signals,bridge_dat[cond]['io_cluster_ids'][run_id], amplitudes = weights_arg)

                within_mean_phase_array = np.stack([within_mean_phase[clust] for clust in within_mean_phase], axis=1)
                mean_amplitude_clusters_array = np.stack([mean_amplitude_clusters[clust] for clust in mean_amplitude_clusters], axis=1)
                weights_arg = mean_amplitude_clusters_array if weight_use == "weighted" else None

                R_K_sync_between, between_mean_phase = Kuramoto_order(within_mean_phase_array, weights = weights_arg)


                # -------------- Kuramoto synchrony of IO-I_OU phase differences -------
                # If all the phase differences are tightly clustered and the Kuramotor order is high, then all the IO's are strongly driven to their OU at a phase locked values. If the phase differences are spread out with a low Kuramoto order, then the IO's are showing behaviour independent of their OU. Do different clusters maintain a similar phase relationship to their OU input, or do they have distinct preferred IO–OU phase lags?

                weights_arg = amplitude_signals if weight_use == "weighted" else None

                OU_sync_global, OU_global_mean_phase = Kuramoto_order(phase_differences, weights =  weights_arg )
                OU_sync_within, OU_within_mean_phase, OU_mean_amplitude = within_cluster_Kuramoto(phase_signals,bridge_dat[cond]['io_cluster_ids'][run_id],  amplitudes =  weights_arg)

                OU_within_mean_phase_array = np.stack([OU_within_mean_phase[clust] for clust in OU_within_mean_phase], axis=1)
                OU_mean_amplitude_array = np.stack([OU_mean_amplitude[clust] for clust in OU_mean_amplitude], axis=1)
                weights_arg = OU_mean_amplitude_array if weight_use == "weighted" else None

                OU_sync_between, OU_between_mean_phase =Kuramoto_order(OU_within_mean_phase_array, weights =  weights_arg)


                # bucket['R_K_sync_global'][] = R_K_sync_global
                # bucket['R_K_sync_within'][] = R_K_sync_within
                # bucket['R_K_sync_between'][] = R_K_sync_between
                # bucket['global_mean_phase'][] = global_mean_phase
                # bucket['within_mean_phase'][] = within_mean_phase
                # bucket['between_mean_phase'][] = between_mean_phase
                # bucket['OU_sync_global'][] = OU_sync_global
                # bucket['OU_sync_within'][] = OU_sync_within
                # bucket['OU_sync_between'][] = OU_sync_between
                # bucket['OU_global_mean_phase'][] = OU_global_mean_phase
                # bucket['OU_within_mean_phase'][] = OU_within_mean_phase
                # bucket['OU_between_mean_phase'][] = OU_between_mean_phase

```

```python
# Get IO-PC synchrony for each IO - PC connection
# Load data
dat_pre = np.load(p_bridges_filepath, allow_pickle=True)
current_dat = {
    cond: dat_pre[cond].item()   # unwrap 0-D object array -> dict
    for cond in ("static", "plasticity")
}
dt = current_dat["static"]['dt'][0]
# Get connectivity measurements
for cond in ("static", "plasticity"):
    network_metrics = calc_network_metrics(network_file_paths= current_dat[cond]['network_graph'])
    current_dat[cond].update(network_metrics)

idx_runs_by_param  = group_runs_by_variable('n_bridges', current_dat)


io_phases_all= {}
io_amplitudes_all = {}
pc_phases_all = {}
pc_amplitudes_all= {}
io_sources= {}
pc_targets = {}

Fs = 1 / (dt / 1000)

for cond in ("static", "plasticity"):
    # Calculate phases
    io_phases_all[cond], io_amplitudes_all[cond]= filter_and_Hilbert(current_dat[cond]['io.V_soma'], Fs)
    pc_phases_all[cond], pc_amplitudes_all[cond]= filter_and_Hilbert(current_dat[cond]['pc.V'], Fs)

    io_sources[cond] = current_dat[cond]["io_pc_pre"]
    pc_targets[cond] = current_dat[cond]["io_pc_post"][:, 0:100]

baseline_duration = 200 # ms
window_duration = 300 #ms

results = nested_dict()

for value, run_ids  in idx_runs_by_param.items():
     for cond in ("static", "plasticity"):
        for run_id in run_ids:
            for con_id in np.arange(n_connections):

                current_pc = pc_targets[cond][run_id, con_id]
                current_io = io_sources[cond][run_id, con_id]

                # cspk_list_unfiltered = np.where(current_dat[cond]['pc.cspk'][run_id, : , current_pc])[0]
                # cspk_list = filter_cspk_list(cspk_list_unfiltered, dt = dt, t_before = baseline_duration, t_after = window_duration, merge_window = 2, overlap= False)

                cspk_list = np.where(current_dat[cond]['pc.cspk'][run_id, : , current_pc])[0]

                for cspk_i, cspk in enumerate(cspk_list):
                    baseline_idx = [cspk-(int(round(baseline_duration /dt))), cspk-(int(round(1/dt)))]
                    window_idx = [cspk+(int(round(20 /dt))), cspk+(int(round(window_duration /dt)))]

                    baseline_IO_phase= io_phases_all[cond][run_id,baseline_idx[0]:baseline_idx[1] , current_io]
                    baseline_IO_amplitude= io_amplitudes_all[cond][run_id,baseline_idx[0]:baseline_idx[1], current_io]
                    baseline_PC_phase=  pc_phases_all[cond][run_id,baseline_idx[0]:baseline_idx[1], current_pc]
                    baseline_PC_amplitude = pc_amplitudes_all[cond][run_id,baseline_idx[0]:baseline_idx[1], current_pc]

                    window_IO_phase = io_phases_all[cond][run_id,window_idx[0]:window_idx[1] , current_io]
                    window_IO_amplitude = io_amplitudes_all[cond][run_id,window_idx[0]:window_idx[1] , current_io]
                    window_PC_phase=  pc_phases_all[cond][run_id,window_idx[0]:window_idx[1] , current_pc]
                    window_PC_amplitude = pc_amplitudes_all[cond][run_id,window_idx[0]:window_idx[1] , current_pc]

                    baseline_d_IO_PC = baseline_IO_phase - baseline_PC_phase
                    window_d_IO_PC = window_IO_phase - window_PC_phase

                    weights_to_use = {'unweighted': (None, None), 'IO_weighted': (baseline_IO_amplitude, window_IO_amplitude) , 'PC_weighted':(baseline_PC_amplitude, window_PC_amplitude)}
                    for weight_use, amplitude_dat in weights_to_use.items():
                        n_connections = len(current_dat['static']["io_pc_pre"])
                        n_runs = len(run_ids)
                        n_cspks = len(cspk_list)
                        n_datapoints_baseline = len(baseline_d_IO_PC)
                        n_datapoints_window = len(window_d_IO_PC)

                        bucket = results[cond][weight_use].setdefault(
                            f'{value}',

                            # Baseline / pre cspk data:
                            {'R_IO_PC_p_con_cspk_pre': np.zeros((n_connections, n_runs, n_cspks,n_datapoints_baseline )),
                            'mean_phase_IO_PC_p_con_cspk_pre':  np.zeros((n_connections, n_runs, n_cspks,  n_datapoints_baseline )),
                             "R_IO_PC_sync_global_pre": np.zeros(n_datapoints_baseline),

                            # Acticity / post cspk window data:
                            'R_IO_PC_p_con_cspk_post': np.zeros((n_connections, n_runs, n_cspks,n_datapoints_baseline )),
                            'mean_phase_IO_PC_p_con_cspk_post':  np.zeros((n_connections, n_runs, n_cspks,  n_datapoints_baseline )),
                             "R_IO_PC_sync_global_post": np.zeros(n_datapoints_baseline),

                             # Baseline subtracted data:
                             "R_IO_PC_p_con_cspk_delta": np.zeros((n_connections, n_runs, n_cspks,n_datapoints_baseline )),
                             "R_IO_PC_sync_global_baselined": np.zeros(n_datapoints_baseline)
                            })

                        # Get baseline window synchrony from period before each incoming IO spike (200 ms) for each IO-PC connection
                        rho_pre, mu_pre = Kuramoto_order( baseline_d_IO_PC, weights = amplitude_dat[0])

                        # Get activity window synchrony from period after each incoming IO spike for each IO-PC connection
                        rho_post, mu_post = Kuramoto_order( window_d_IO_PC, weights = amplitude_dat[1])

                        # Baseline subtracted
                        rho_baseline_subt = rho_post - rho_pre

                        # opulation-level, cspk-triggered synchrony measure across all IO–PC connections


                        bucket['R_IO_PC_p_con_cspk_pre'][con_id, run_id, cspk_i,:] = rho_pre
                        bucket['mean_phase_IO_PC_p_con_cspk_pre'][con_id, run_id, cspk_i,:] = mu_pre
                        bucket['R_IO_PC_p_con_cspk_post'][con_id, run_id, cspk_i,:] = rho_post
                        bucket['mean_phase_IO_PC_p_con_cspk_post'][con_id, run_id, cspk_i,:] = mu_post
                        bucket["R_IO_PC_p_con_cspk_delta"][con_id, run_id, cspk_i,:] = rho_baseline_subt


# Get baseline window synchrony from period before each incoming IO spike (200 ms) for each IO-PC connection

# Get activity window synchrony from period after each incoming IO spike for each IO-PC connection

# Merge all cpsks and all connections

# Show in plots + where delay from IO to PC takes place and where loop delay

# Subtract activity-baseline to get Cspk triggered synchrony for all IO-PC connections per parameter
```
